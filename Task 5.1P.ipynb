* **Load the dataset from “sampleData.csv”.  Create training and test datasets using random splitting (70-30)%. Print the train and test data. Create a linear regression model and report the model performance. (The last feature is the target variable)**

```python
#Load data
df = pd.read_csv('sample_dataset.csv')
```

```python
#Get the features and labels
X = df.iloc[:,:-1]
y = df.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
#Split train and test dataset 7:3
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
y_train.head()
```

```python
X_test.head()
```

```python
y_test.head()
```

```python
#Build linear model
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
pred = lr.predict(X_test)
```

```python
#Validate the model
from sklearn.metrics import mean_squared_error, mean_absolute_error, confusion_matrix, classification_report
```

```python
print('Mean Square Error: ', round(mean_squared_error(y_test, pred),2))
print('Mean Absolute Error: ', round(mean_absolute_error(y_test, pred),2))
print('Classification report: ', classification_report(y_test.round(), pred.round()))
```

* **Create training and test dataset using leave-one-out based on Subject ID. Print the train and test dataset for the first iteration only.  Create a linear regression model and compare the performance with Q-1. Explain your findings.**

```python
from sklearn.model_selection import cross_val_score, LeaveOneOut, LeaveOneGroupOut
```

```python
X = df.iloc[:,:-1].values
y = df.iloc[:, -1].values
```

```python
logo = LeaveOneGroupOut()
```

```python
groups = df['Subject ID']

y_true, y_pred = list(), list()

for i, (train_index, test_index) in enumerate(logo.split(X, y, groups)):
    
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    if i == 1:
        print('X_train: ', X_train)
        print('X_test: ', X_test)
        print('y_train: ', y_train)
        print('y_test: ', y_test)
        
    lr_model = LinearRegression()
    lr_model.fit(X_train, y_train)
    
    predict = lr_model.predict(X_test)
    
    y_true.append(y_test[0])
    y_pred.append(predict[0])
    
print('Mean Square Error: ', round(mean_squared_error(y_true, y_pred),2))
print('Mean Absolute Error: ', round(mean_absolute_error(y_true, y_pred),2))
print('Classification report: ', classification_report(np.around(np.array(y_true)), np.around(np.array(y_pred))))    
```

As we can see that all the metrics from the train test split and the leave one group out method are pretty the same: mean square error is around 0.12, mean absolute error is 0.3, accuracy is around 50%.
