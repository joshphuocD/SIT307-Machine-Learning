```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

* **Read "train_wbcd.csv" and print the feature name with numbers of missing entries.**


```python
#Read data file
df = pd.read_csv('train_wbcd.csv')
```

```python
#Print the first 5 rows
df.head()
```

```python
#Get an overview of the data
df.describe()
```

```python
#Print missing values
df.isnull().sum()
```

* **Fill in the missing entries.
For filling any feature, you can use either the mean or median value of the feature values from observed entries.**

```python
#caculate mean value
mean_val = df['f21'].mean()
#fill missing values by mean
df['f21'] = df['f21'].fillna(mean_val)
```

```python
#Recheck null data
df.isnull().sum()
```

* **Explain the reason behind your choice and print replacement value of each feature.**

I chose to replace the missing values by the mean because it represents the average value of the data set

* **Encode the categorical variable "Diagnosis" using an appropriate encoding approach and display the encoded values. Justify your selection of encoding approach. Save the modified data into a csv file**

```python
df.head()
```

```python
#Create dummies values for Diagnosis B and M. 1 is positive and 0 is negative
df = pd.get_dummies(df, prefix=['Diagnosis'], columns = ['Diagnosis'])
df.head()
```

```python
#Save the modified data into a csv file
df.to_csv('train_wbcd_dummies.csv')
```

* **Apply the min-max scaling on features f1 to f30. Plot distribution of first six features before and after scaling. Is there any difference? Please explain.**

```python
#Plot the first 6 features before scaling
labels = df.columns[1:7]
plt.plot(df.iloc[:, 1:7])
plt.legend(labels,loc='upper left')
plt.show()
```

*As we can see that one feature is extremely higher than the rest, it is not easy to visualize and analyse*

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(df)
```

```python
scaled_df = pd.DataFrame(scaler.transform(df))
```

```python
scaled_df.head()
```

```python
plt.plot(scaled_df.iloc[:,1:7])
plt.legend(labels, loc ='upper left')
plt.show()
```

*After being scaled, the features are now from 0 to 1 and possible to be visualized altogether*
